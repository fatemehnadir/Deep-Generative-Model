## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Read data

```python
df = pd.read_csv('ngaussian.csv', header=None) 
df[['X1', 'X2']] = df[0].str.split(' ', expand=True)
df = df.drop(columns=[0])
```

# Show data

```python
df.head()
```

```python
X1 = np.array(df['X1'].values.reshape(-1,1),dtype=float)
X2 = np.array(df['X2'].values.reshape(-1,1),dtype=float)
```

## Part a

```python
reg = LinearRegression().fit(X1, X2)
beta_12 = reg.coef_[0]
print("beta_12:", beta_12)
```

```python
beta_12 = beta_12.reshape(-1,1)
resid = X2 - beta_12 * X1
```

```python
plt.scatter(X1, resid)
plt.xlabel('X1') 
plt.ylabel('Residual')
plt.title('Problem 5(a)')
plt.show()
```
