```python
import matplotlib.pyplot as plt
from datasets import load_dataset

def plot_sequence_lengths(dataset):

    # Initialize a list to store the sequence lengths
    sequence_lengths = []

    # list of indices that are too long
    too_long = []

    # Loop over the dataset and get the lengths of text sequences
    for idx, example in enumerate(dataset["train"]):
        sequence_lengths.append(len(example['inputs']))
        if sequence_lengths[idx] > 2200:
          too_long.append(idx)

    # Plot the histogram
    plt.hist(sequence_lengths, bins=30)
    plt.xlabel('Sequence Length')
    plt.ylabel('Count')
    plt.title('Distribution of Text Sequence Lengths')
    plt.show()

    return too_long
```

Clean text

```python
def clean_text(text):
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@[^\s]+", "", text)
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r'[^a-zA-Z0-9,!.?<> ]', '', text) #
    text = re.sub(r'(\W)(?=\1)', '', text) #
    return re.sub(r"\^[^ ]+", "", text)
```

Make the structure of two datasets similar

```python
def clean_statement_train(data_point):
    statement = data_point["inputs"]
    
    if "Statement:" in statement and "Plausible/implausible?" in statement:
        
        cleaned_statement = statement.split("Statement:")[1].split("Plausible/implausible?")[0].strip()
        final_sentence = f"Is the following sentence plausible?\n\"{cleaned_statement}.\""

        data_point['inputs'] = clean_text(final_sentence)
        data_point['targets'] = 'yes' if data_point['targets'][0].startswith('p') else 'no'
        return data_point
    else:
        return "Invalid statement format"
```

```python
def clean_statement_test(data_point):

    data_point['input'] = clean_text(data_point['input']).strip()
    return data_point
```

```python
def process_dataset(data: Dataset, mode):
  if mode == "train":
    return (
        data.shuffle(seed=42)
        .map(clean_statement_train)
        .remove_columns(
            [
                "multiple_choice_targets",
                "multiple_choice_scores",
                "idx"
            ]
        )
    )
  else:
    return (
      data.shuffle(seed=42)
      .map(clean_statement_test)
  )

```

```python
import pandas as pd
from tqdm.notebook import tqdm

columns = ['prompt', 'original_target', "model_target","generated_text"]
df_zeroshot = pd.DataFrame(columns=columns)


prompt_template = """
# Answer the following yes/no question.
{input}
"""


texts = [
    prompt_template.format(
        input=row['input']
    )
    for row in dataset['test']
]
summaries = [
    row['target']
    for row in dataset['test']
]

model.config.pad_token_id = model.config.eos_token_id

generated_text = []
prompts = []

for prompt in tqdm(texts):

  inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
  tokens = model.generate(
    **inputs,
    max_new_tokens=1,
    temperature=0.00001,
    do_sample=True,
    pad_token_id=model.config.eos_token_id,  # Set pad_token_id directly in the generate method
    eos_token_id=model.config.eos_token_id,  # Set eos_token_id directly in the generate method

  )

  generated_text.append((tokenizer.decode(tokens[0], skip_special_tokens=True)))
  prompts.append(prompt)
```

```python
import re
import random

def find_last_yes_or_no(s):
    matches = re.findall(r'\b(?:yes|no)\b', s, flags=re.IGNORECASE)
    if matches:
        return matches[-1]
    else:
        return random.choice(["yes", "no"])
res = []
for sgen in generated_text:
  res.append(find_last_yes_or_no(sgen))
```

```python
df_zeroshot['prompt'] = prompts
df_zeroshot['original_target'] = summaries
df_zeroshot['model_target'] = res
df_zeroshot['generated_text'] = generated_text
df_zeroshot.to_csv("/content/drive/MyDrive/MSC/Fall2023/DGM/HW3/Q2/zeroshot.csv")
```

# 3-Shot

```python
def make_prompt(example_indices_full, example_index_to_summarize):
    # prompt = ''
    for index in example_indices_full:
        dialogue1 = dataset['train'][index]['inputs']
        summary1 = dataset['train'][index]['targets']
        dialogue2 = dataset['train'][index+2]['inputs']
        summary2 = dataset['train'][index+2]['targets']
        dialogue3 = dataset['train'][index+4]['inputs']
        summary3 = dataset['train'][index+4]['targets']
        dialoguee = dataset['test'][example_index_to_summarize]['input']

        prompt = ""

        prompt += f"""# Q: Answer the following yes/no question.
{dialogue1}
# A: {summary1}

# Q: Answer the following yes/no question.
{dialogue2}
# A: {summary2}

# Q: Answer the following yes/no question.
{dialogue3}
# A: {summary3}

# Q: Answer the following yes/no question.
{dialoguee}
# A:
"""

    return prompt
```

```python
example_indices_full = [1]
example_index_to_summarize = 40

three_shot_prompt = make_prompt(example_indices_full, example_index_to_summarize)

print(three_shot_prompt)
```

```python
target = dataset['test'][example_index_to_summarize]['target']

inputs = tokenizer(three_shot_prompt, return_tensors='pt').to("cuda")
output = tokenizer.decode(
    model.generate(
        inputs["input_ids"],
        max_new_tokens=1,
        temperature=0.00001,
        do_sample=True,
    )[0],
    skip_special_tokens=True
)

print(dash_line)
print(f'BASELINE HUMAN TARGET:\n{target}\n')
print(dash_line)
print(f'MODEL GENERATION - 3 SHOT:\n{output}')
```
